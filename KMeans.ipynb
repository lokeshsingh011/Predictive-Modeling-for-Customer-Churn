```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize']=(16,9)
plt.style.use('ggplot')
```

```python
data=pd.read_csv('mallCustomerData.txt', sep=",")
print(data.shape)
data.head(10)
```

```python
print(data['Gender'].value_counts())
```

```python
f1=data['Annual Income (k$)'].values
f2=data['Spending Score (1-100)'].values

for key in data.keys():
    print(key)

X=np.array(list(zip(f1,f2)))
plt.scatter(f1,f2, c='black', s=20)
```

```python
from sklearn.cluster import KMeans

km=KMeans(n_clusters=3)
km=km.fit(X)
labels=km.predict(X)
c=km.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,1])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')  # Create 3D axes in the figure
ax.scatter(X[:, 0], X[:, 1], X[:, 1])  # Assuming z-values are in the 3rd column
plt.show()  # This line is necessary to display the plot
```
